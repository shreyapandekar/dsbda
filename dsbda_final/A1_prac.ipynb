```python
df["Total Marks"]=df["Total Marks"].astype("float")
```

```python
df['Gender'].replace({"M":0,"F":1},inplace=True)
```

```python
#define a mapping dictionary for categorical to quantitative values
mapping = {'Comp':1, 'IT':2, 'ENTC':3}
#map the 'origin' column using the defined mapping
df['Branch_encoded'] = df['Branch'].map(mapping)

df.head()
```

```python
#Label encoding method
df["Attendence"]=df["Attendence"].astype("category")
df["Attendence"]=df["Attendence"].cat.codes
```

```python
df
```

```python
df["Phy_marks"].fillna(df["Phy_marks"].mean(),inplace=True)
```

```python
df["Che_marks"].fillna(df["Che_marks"].mean(),inplace=True)
```

```python
df['Che_marks'] = df['Che_marks'].fillna(np.mean(df['Che_marks']))
df
```

```python
df["Phy_marks"]=df["Phy_marks"].fillna(np.mean(df["Phy_marks"]))
df
```

```python
df["SME_marks"]=(df["SME_marks"]-df["SME_marks"].min())/(df["SME_marks"].max()-df["SME_marks"].min())
```
