```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
print('Information of Dataset:\n', titanic.info)
print('Shape of Dataset (row x column): ', titanic.shape)
print('Columns Name: ', titanic.columns)
print('Total elements in dataset:', titanic.size)
print('Datatype of attributes (columns):', titanic.dtypes)
print('First 5 rows:\n', titanic.head())
print('Last 5 rows:\n',titanic.tail().T)
print('Any 5 rows:\n',titanic.sample(5).T)
```

```python
titanic.isna().sum()
```

```python
titanic['age'] = titanic['age'].fillna(np.mean(titanic['age']))
```

```python
titanic['embarked']=titanic['embarked'].replace({"S":1,"C":2,"Q":3})
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

```python
titanic.head()
```

```python
titanic['deck'] = titanic['deck'].fillna(titanic['deck'].mode()[0])
```

```python
titanic["deck"]=titanic["deck"].fillna(titanic["deck"].mode()[0])
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna(titanic['embark_town'].mode()[0])
```

```python
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
```

```python
# 2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is
# distributed by plotting a histogram

sns.histplot(data=titanic, x='fare', kde=True, bins=16)
plt.title('Distribution of Fares')  # Set the title of the plot
plt.show()
```

```python

# Count the number of survivors and non-survivors for each gender
male_survivors = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 1)].shape[0]
male_non_survivors = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 0)].shape[0]
female_survivors = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)].shape[0]
female_non_survivors = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 0)].shape[0]

print(f"Male survivors: {male_survivors}")
print(f"Male non-survivors: {male_non_survivors}")
print(f"Female survivors: {female_survivors}")
print(f"Female non-survivors: {female_non_survivors}")
```

```python

# Plot a box plot of the 'age' column with respect to 'sex' and 'survived'
plt.figure(figsize=(10,6))  # Set the figure size
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Box Plot of Age Distribution by Sex and Survival')  # Set the title of the plot
plt.show()
```

```python
fig, axis = plt.subplots(1,3)
sns.histplot(ax = axis[0], data = titanic, x='sex', hue = 'sex')
sns.histplot(ax = axis[1], data = titanic, x='pclass', hue = 'pclass',multiple = 'dodge', shrink = 0.8)
sns.histplot(ax = axis[2], data = titanic, x='survived', hue = 'survived', multiple = 'dodge', shrink = 0.8)
plt.show()
```

```python


# Two variable histogram
fig, axis = plt.subplots(2,2)
sns.histplot(ax = axis[0,0], data = titanic, x='age', hue = 'sex', multiple = 'dodge', shrink = 0.8, kde = True)
sns.histplot(ax = axis[0,1], data = titanic, x='fare', hue = 'sex', multiple = 'dodge', shrink = 0.8, kde = True)
sns.histplot(ax=axis[1,0], data=titanic, x='age', hue = 'survived', multiple = 'dodge',shrink=0.8, kde= True)
sns.histplot(ax = axis[1,1], data=titanic, x='fare', hue='survived', multiple='dodge', shrink=0.8, kde = True)
plt.show()
```
