```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=sns.load_dataset("titanic")
```

```python
dataset
```

```python
print('Information of Dataset:\n', dataset.info)
print('Shape of Dataset (row x column): ', dataset.shape)
print('Columns Name: ', dataset.columns)
print('Total elements in dataset:', dataset.size)
print('Datatype of attributes (columns):', dataset.dtypes)
print('First 5 rows:\n', dataset.head().T)
print('Last 5 rows:\n',dataset.tail().T)
print('Any 5 rows:\n',dataset.sample(5).T)
```

```python
#---------------------------------------------------------------------------------------
#One variable
fig, axes = plt.subplots(1,2)
sns.boxplot(data = dataset, y ='age', ax=axes[0])
sns.boxplot(data = dataset, y ='fare', ax=axes[1])
plt.show()
```

```python
# Two variables
fig, axes = plt.subplots(1,3)
sns.boxplot(data = dataset, x='sex', y ='age', hue = 'sex', ax=axes[0])
sns.boxplot(data = dataset, x='pclass', y ='age', hue = 'pclass', ax=axes[1])
sns.boxplot(data = dataset, x='survived', y ='age', hue = 'survived', ax=axes[2])
plt.show()
```

```python
sns.boxplot(x='sex',y='age',data=dataset)
```

```python
sns.boxplot(x='sex',y='age',data=dataset,hue='survived')
```

```python
#three variables
fig, axes = plt.subplots(1,2, sharey=True)
sns.boxplot(data = dataset, x='sex', y ='age', hue = 'survived', ax=axes[0])
sns.boxplot(data = dataset, x='pclass', y ='age', hue = 'survived', ax=axes[1])
plt.show()
```

```python
sns.histplot(data=dataset['age'],bins=8)
```

```python
sns.histplot(data=dataset['sex'],bins=8)
```

```python
# Display the distribution of 'sex' and 'class'
sns.countplot(x='sex', data=dataset)
plt.show()

sns.countplot(x='class', data=dataset)
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=dataset)
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=dataset)
plt.show()
```

```python
sns.barplot(x="class", y="survived", data=dataset)
```

```python
sns.barplot(x="sex",y='age',hue="survived",data=dataset)
```
