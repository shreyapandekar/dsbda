```python
def DetectOutliers(df,var):
    Q1 = df['Che_marks'].quantile(0.25)
    Q3 = df['Che_marks'].quantile(0.75)
    IQR = Q3 - Q1

    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    print("Highest allowed in vairable: ",var,high)
    print("Lowest allowed in vairable: ",var,low)

    count = df[(df[var]>high) | (df[var] < low)][var].count()
    print("Total outliers in : ",var,':',count)

    df1 = df[((df[var]<low) | (df[var]>high))]
    print("Outliers : \n",len(df1))
    print(df1.T)


    df = df[((df[var] >= low) & (df[var] <= high))]
    return(df)

```

```python
df=DetectOutliers(df,'Che_marks')
```

```python
fig,axes = plt.subplots(2,2)
fig.suptitle('After remocing Outliers')
sns.boxplot(data = df,x = 'Phy_marks',ax=axes[0,0])
sns.boxplot(data = df,x = 'Che_marks',ax=axes[0,1])
sns.boxplot(data = df,x = 'EM1_marks',ax=axes[1,0])
sns.boxplot(data = df,x = 'SME_marks',ax=axes[1,1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some non-linear data
x = np.linspace(1, 10, 100)
y = np.log(x) + np.random.normal(0, 0.2, 100)

# Plot the original data
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.show()

# Apply logarithmic transformation
x_log = np.log(x)
y_log = np.log(y)

# Plot the transformed data
plt.scatter(x_log, y_log)
plt.xlabel('log(X)')
plt.ylabel('log(Y)')
plt.title('Linearized Relationship')
plt.show()
```
